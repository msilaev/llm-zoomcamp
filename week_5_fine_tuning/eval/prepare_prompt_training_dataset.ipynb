## Prepare prompt-style fine tuning training dataset

```python
import json
import random
from pathlib import Path
```

```python
# -------- CONFIG --------
PARENT_DIR = Path("C:/Users/mikes/Documents/STUDY/llm_fine_tuning_data")
DATA_FILE = PARENT_DIR / "raw_qa_merged.jsonl"

TRAIN_FILE = PARENT_DIR / "train.jsonl"
VAL_FILE = PARENT_DIR / "val.jsonl"
VAL_SPLIT = 0.1
# ------------------------
```

```python

SYSTEM_PROMPT = (
    "You are a knowledgeable machine learning assistant. "
    "Answer questions clearly, accurately, and concisely."
)

def format_example(q, a):
    return {
        "messages": [
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": q},
            {"role": "assistant", "content": a}
        ]
    }

with open(DATA_FILE) as f:
    data = [json.loads(line) for line in f]

random.shuffle(data)
split = int(len(data) * (1 - VAL_SPLIT))

train, val = data[:split], data[split:]

with open(TRAIN_FILE, "w") as f:
    for x in train:
        f.write(json.dumps(format_example(x["question"], x["answer"])) + "\n")

with open(VAL_FILE, "w") as f:
    for x in val:
        f.write(json.dumps(format_example(x["question"], x["answer"])) + "\n")

print(f"Train: {len(train)}, Val: {len(val)}")

```
